#Q1) Write a program to request a file name from the user and calculate the
following statistics of the contents of the file:
 Number of lines
 Number of words
 Number of characters
 Average length of a word

**Explanation:**

We first ask the user for the filename using the input() function and store it in the variable filename.

We then use the with statement to open the file in read mode and create a file object file.

We use the readlines() method of the file object to read all the lines of the file and store them in a list called lines. We then find the number of lines by taking the length of the lines list.

We initialize the variables num_words and num_chars to 0. We then loop through each line in the lines list and use the split() method to split the line into words. We find the number of words in each line by taking the length of the words list and add it to num_words. We also loop through each word in the words list and find its length using the len() function. We add the length of each word to num_chars.

We find the average length of a word by dividing num_chars by num_words and store it in the variable avg_word_length.

Finally, we print the results using the print() function. We use f-strings to format the output with the appropriate labels and values. We use the :.1f format specifier to round the average word length to one decimal place.

```python
filename = input("What is the filename: ")

with open(filename, 'r') as file:
    lines = file.readlines()
    num_lines = len(lines)
    
    num_words = 0
    num_chars = 0
    
    for line in lines:
        words = line.split()
        num_words += len(words)
        for word in words:
            num_chars += len(word)
    
    avg_word_length = num_chars / num_words
    
    print(f"Number of lines: {num_lines}")
    print(f"Number of words: {num_words}")
    print(f"Number of characters: {num_chars}")
    print(f"Average length of a word: {avg_word_length:.1f}")
```

#Q2) A string is an anagram of another if the second string is simply a scrambled

version of the first. Write a python program to implement the following game:
a) Reads in a file that has words and their meanings in a text file. An example
“words and their meanings” file is given in canvas. Note that your program
needs to ask the user for the “words and their meanings” file to use.
b) The words and their meanings text file is of the csv (comma separated values)
format. Use either notepad++ or notepad to create your file in the same format
as the mywords.txt file given to you in canvas
c) Your program should then pick a word from the “words and their meanings”
file, scramble the letters, and ask the user to unscramble it. Every run of your
program should pick a word at random.
d) The user may type in the unscrambled word or may ask for the
definition/meaning of the word by entering a question mark.
e) The game continues for the number of times as the number of letters in the
word. For example if the word is “poor”, the program will allow four
attempts.
f) The program will also provide the definition of the word (at the user’s
request), only once. If the user enters “?” more than once, an error message
will be given with the warning that the next input of a “?” will be counted as
an attempt at the answer.
g) until the user says “no” to the question: “Do you want to continue?”. A
sample run is shown below (user inputs are in red):
Give the name of the “words and their meanings” file: mywords.txt
Do you want to play: Y
Unscramble the following letters to form a word. Type “?” for the meaning
of the unscrambled word: ulecenop
Enter the answer [or ? for the meaning]: ?
The word means: great wealth
Enter the answer [or ? for the meaning]: rich
Wrong, try again

**explanation:**

The program asks the user for the name of the "words and their meanings" file to use.
The program reads in the contents of the file and stores them in a dictionary where the keys are the words and the values are their meanings.
The program prompts the user to play the game and starts a loop that runs until the user says they don't want to play anymore.
Inside the loop, the program selects a random word from the dictionary, scrambles the letters of the word, and presents it to the user.
The program then prompts the user to enter their answer or to request the meaning of the word by typing a "?".
If the user types a "?", the program prints the meaning of the word and increments a counter that keeps track of the number of times the meaning has been requested.
If the user types an answer, the program checks if it is correct. If it is, the program congratulates the user and breaks out of the loop.
If the user types an incorrect answer, the program prompts them to try again until they have reached the maximum number of attempts.
After each guess, the program prints a message that informs the user of how many attempts they have left.
If the user has used up all their attempts, the program prints a message and breaks out of the loop.
After the game is finished, the program prompts the user if they want to play again.
If the user says they don't want to play again, the program prints a goodbye message and exits.

```python
import csv
import random

# Function to read in the words and their meanings file and store it in a dictionary
def read_words_file(filename):
    words_dict = {}
    with open(filename, 'r') as csvfile:
        reader = csv.reader(csvfile)
        for row in reader:
            words_dict[row[0]] = row[1]
    return words_dict

# Function to scramble a word
def scramble_word(word):
    letters = list(word)
    random.shuffle(letters)
    return ''.join(letters)

# Function to get user input and check if it is a valid answer or a request for the word's meaning
def get_user_input(word, words_dict):
    while True:
        user_input = input("Enter the answer [or ? for the meaning]: ").strip().lower()
        if user_input == "?":
            if word in words_dict:
                print("The word means:", words_dict[word])
            else:
                print("Sorry, the meaning of this word is not available.")
        elif len(user_input) == len(word):
            return user_input
        else:
            print("Invalid input! Please enter a valid answer or '?' for the meaning.")

# Main program
filename = input("Give the name of the \"words and their meanings\" file: ")
words_dict = read_words_file(filename)

play = input("Do you want to play? (Y/N) ").strip().lower()
while play == "y":
    # Select a random word from the dictionary
    word = random.choice(list(words_dict.keys()))
    scrambled_word = scramble_word(word)
    print("Unscramble the following letters to form a word. Type '?' for the meaning of the unscrambled word:")
    print(scrambled_word)

    # Play the game
    max_attempts = len(word)
    attempts = 0
    has_asked_for_meaning = False
    while attempts < max_attempts:
        answer = get_user_input(word, words_dict)
        if answer == word:
            print("You got it!")
            break
        else:
            attempts += 1
            if attempts == max_attempts:
                print("Wrong, you have exceeded the number of attempts. Bye!")
            else:
                print("Wrong, try again")

            # Check if the user has asked for the word's meaning before
            if answer == "?" and has_asked_for_meaning:
                attempts += 1
                print("You have been given the meaning before. Next time you ask for the meaning it will count as an attempt!")
            elif answer == "?":
                has_asked_for_meaning = True

    # Ask the user if they want to continue playing
    play = input("Do you want to play? (Y/N) ").strip().lower()

print("Goodbye!")
```

# Q3) Cosine Similarity


The  file  poetry_lines.txt  is  given  to  you. The file contains names of poets and an extract of their poetry. New lines in each poem are represented by a ‘/’. 

The format of a line is the following: <Poet’s name>:<poetry delineated by ‘/’>\n 

The first string in each line is the name of the author followed by a ‘:’, followed by the poetry which is delineated by ‘/’ to represent a new line in the poem. The next line contains the next poem and so on. 
 
You  are  required  to  input  a  few  lines  of  your  own  poem  to  the  python  program (with lines separated by “/”)  and  compute  the  cosine  distance  (similarity  score) between  each  line  (of  poetry  from  the  file)  and  your  own  poem.  Finally  your program should display the following: 
1.   Each poet and the similarity score with your poem.
2.   Finally display the poem that is closest to your input.


 
An  example  of  how  to  compute  the  cosine  distance  between  two  lines  of  text  is given below: 
 
Line1: Hi Hi Hi how are you 

Line2: Hi how are u u 
 
* The cosine similarity between two vectors (A and B) is given by the formula:



 > Similarity = Cos(0) = (A.B) / ||A||.||B||

* Create a dictionary of keys with the words and the values with the number of 
occurrences.  
> 1. Line1dict = {‘Hi’:3,’how’:1,’are’:1,’you’:1,’u’:0} 
> 2.  Line2dict = {{‘Hi’:1,’how’:1,’are’:1,’you’:0,’u’:2} 
* The  values  in  the  two  dictionaries  form  the  two  vectors  to  be  compared. For example: 
> 1.  Line1vector = (3,1,1,1,0) 
> 2.  Line2vector = (1,1,1,0,2) 
* Now  you  can  implement  the  cosine  similarity  formula  above  to  compare Line1vector and Line2vector. 
*  Note that the cosine distance should be between 0 and 1. 
 

Your program should provide the following: 
1.  Ability to read in a user provided filename which contains names of poets and an extract of their poetry in the specified format. 
2.  Read in user’s own poem (using the input() method). 
3.  Functionality to compute the similarity (use cosine distant) between each poem and the user’s poem and display the results. 
4.  Finally choose and display the poem that is most similar to the user’s input.


**Explanations:**
> 1. Read file by path provided by user.
> 2. Parse file in a dictionary of nested list, where list would hold poetry lines.
> 3. Calculate word count vector for each line for every poet.
> 4. Read user input poetry.
> 5. Calculate word count vector for each line of user input.
> 6. Compare user count vector with corresponding poet and calculate cosine similarity value.
> 7. Display poetry that is most similar to user input.

```python
import re
import math
import copy
import statistics

# Function to read file
def get_input_file():

  try:

    file_path = input('Give the name of the poetry file: ')
    file_object = open(file_path, 'r')

    return file_object

  except:
    print('Invalid file path or file does not exists !')

    return 0


# Function to get word count vector for each line
def get_word_count_vector(line_1, line_2):
  
  if line_1 and line_2:

    line_1 = re.sub(r"[^a-zA-Z.\d\s]", "", line_1)
    line_2 = re.sub(r"[^a-zA-Z.\d\s]", "", line_2)

    vector_1 = []
    vector_2 = []
    
    unique_words = set(line_1.split() + line_2.split())

    for word in unique_words:
      vector_1.append(line_1.count(word))
      vector_2.append(line_2.count(word))

    return vector_1, vector_2

  elif line_1 and not line_2:

    line_1 = re.sub(r"[^a-zA-Z.\d\s]", "", line_1)

    vector_1 = []
    vector_2 = []
    
    unique_words = set(line_1.split())

    for word in unique_words:
      vector_1.append(line_1.count(word))
      vector_2.append(0)

    return vector_1, vector_2

  elif line_2 and not line_1:

    line_2 = re.sub(r"[^a-zA-Z.\d\s]", "", line_2)

    vector_1 = []
    vector_2 = []
    
    unique_words = set(line_2.split())

    for word in unique_words:
      vector_2.append(line_2.count(word))
      vector_1.append(0)

    return vector_1, vector_2

# Function to calculate cosine simaliraty
def calculate_cosine_similaraty(line_1, line_2):

  vector_1, vector_2 = get_word_count_vector(line_1, line_2)

  #print(f'line_1 : {line_1} - vec_1 : {vector_1}')
  #print(f'line_2 : {line_2} - vec_2 : {vector_2}')

  numerator = sum([vector_1[i]*vector_2[i] for i in range(len(vector_1))])
  dennomenator = (math.sqrt(sum([x**2 for x in vector_1])) * math.sqrt(sum([y**2 for y in vector_2])))

  return (numerator/dennomenator) if dennomenator != 0 else 0


# Function to prepare word count vector
def get_similarity(user_poem, poets_poem):

  max_len = max(len(user_poem), len(poets_poem))

  similarity_list = []

  for i in range(max_len):

    line_1 = user_poem.pop(0) if len(user_poem) > 0 else None
    line_2 = poets_poem.pop(0) if len(poets_poem) > 0 else None

    similarity_list.append(calculate_cosine_similaraty(line_1, line_2))

  return similarity_list
```

```python
# Read file
valid_file = get_input_file()

while valid_file == 0:

  print('Enter a valid file path')
  valid_file = get_input_file()

poetry_dict = {}
max_similarity = -1
max_similar_poet = ''

for poem in valid_file:
  print(poem)
  split_poem = poem.strip().split(':')
  if len(split_poem) != 2:
    continue
  key,value = split_poem
  poetry_dict[key] = value.strip().split('/')

# Read user input
user_input_poetry = input('Input your poem delineated by “/”  for  each  line: ').split('/')

print('\nCosine Distance Results:')

for poet,poem_lines in poetry_dict.items():

  similarity = statistics.mean(get_similarity(copy.deepcopy(user_input_poetry), copy.deepcopy(poem_lines)))
  print(f'{poet} : {similarity}')

  if max_similarity < similarity:
    max_similarity = similarity
    max_similar_poet = poet
  
print('The closest poem is:')
print(f'{max_similar_poet} : {"/".join(poetry_dict[max_similar_poet])}')
```

# Q4) Web Scrapper
This is a project to scrape data from the web and store the results in a 
text file and the SQLite database. 

The website https://finance.yahoo.com/trending-tickers lists extensive finance 
data. You have to write Python scripts/programs to collect the current prices 
for the following commodities: Crude Oil, Gold and Silver. Your program 
should store the commodity name and its corresponding price in a text file 
called commodity_prices.txt.

In addition to the commodity_prices.txt file, the data should also be stored in 
an SQLite database called CommodityDatabase in the directory that your 
Jupyter Notebook code will be executed from. The CommodityDatabase 
should have a table called CommodityTable that contains the following 
columns and types: 


> Ticker  :  TEXT

> Price  : REAL

Every execution of your program should create a new commodity_prices.txt 
and CommodityDatabase.db file in the directory (delete any existing files that 
you will create) that your Python script is located and run. 



**Explanations:**

First, we import the necessary modules: requests for making HTTP requests, sqlite3 for interacting with the SQLite database, os for deleting existing files, and BeautifulSoup for parsing HTML.

We define the get_connection() function that returns a connection to the SQLite database. This function is used to create a connection object to the CommodityDatabase.db.

The create_table() function is defined that creates the CommodityTable in the SQLite database. This table has two columns: Ticker (a string) and Price (a floating-point number).

The insert_into_table() function is defined that inserts a commodity name and its price into the CommodityTable.

The fetch_data() function is defined that fetches all the rows from the CommodityTable and prints them.

The get_commodity_price() function is defined that extracts the name and price of a commodity from the BeautifulSoup object.

The code checks whether commodity_prices.txt and CommodityDatabase.db exist in the current directory and deletes them if they do.

The create_table() function is called to create the CommodityTable in the SQLite database.

We use the requests module to fetch the HTML content of the URL https://finance.yahoo.com/trending-tickers.

We use BeautifulSoup to parse the HTML content and extract the commodity names and their prices. We store the commodity names and their prices in a list called commodity_list.

We insert each commodity name and its price into the CommodityTable using the insert_into_table() function.

We write the commodity names and their prices to the commodity_prices.txt file.

Finally, we call the fetch_data() function to fetch all the rows from the CommodityTable and print them.

This program extracts the current prices of Crude Oil, Gold and Silver from finance.yahoo.com and stores the data in a text file (commodity_prices.txt) and SQLite database (CommodityDatabase.db).









```python
import requests
import sqlite3
import os
from bs4 import BeautifulSoup

# Function to create SQLite connection
def get_connection():

    DB_NAME = "CommodityDatabase.db"
    connection = sqlite3.connect(DB_NAME)

    return connection

# Function to create Table
def create_table():

    query = """ CREATE TABLE CommodityTable(
        Id INTEGER PRIMARY KEY AUTOINCREMENT,
        Ticker TEXT,
        Price REAL
    )"""

    con = get_connection()
    con.execute(query)
    con.close()

# Function to insert into table
def insert_into_table(ticker, price):

    query = f"INSERT INTO CommodityTable(Ticker, Price) VALUES('{ticker}', {price})"

    con = get_connection()
    con.execute(query)
    con.commit()
    con.close()

# Function to fetch data from database
def fetch_data():

    query = "SELECT * FROM CommodityTable"

    con = get_connection()
    cur = con.cursor()
    cur.execute(query)

    result = cur.fetchall()

    print(result)

# Function to get commodity price from BeautifulSoup
def get_commodity_price(block):

    commodity_name_block = block.findNext('a', attrs={"class": 'Fz(s) Ell Fw(600) C($linkColor)'})
    commodity_price_block = block.findNext('fin-streamer', attrs={"data-field": "regularMarketPrice"})

    if commodity_name_block is not None and commodity_price_block is not None:
        commodity_name = commodity_name_block.get_text()
        commodity_price = commodity_price_block['value']

        return (commodity_name, commodity_price)
    else:
        return None

# Delete existing files
if os.path.exists('commodity_prices.txt'):
    os.remove('commodity_prices.txt')

if os.path.exists('CommodityDatabase.db'):
    os.remove('CommodityDatabase.db')

# initialize database and create table
create_table()

# Fetch HTML Page
WEB_URL = "https://finance.yahoo.com/trending-tickers"
response = requests.get(WEB_URL)
soup = BeautifulSoup(response.text, 'html.parser')

commodity_list = []

parent_block = soup.find('ul', attrs={"class": "Carousel-Slider Pos(r) Whs(nw)"}).find_all('li')

for children in parent_block:
    commodity_data = get_commodity_price(children)
    if commodity_data is not None:
        commodity_name, commodity_price = commodity_data
        commodity_list.append(f'{commodity_name} : {commodity_price}')
        insert_into_table(commodity_name, commodity_price)

# Write to file
file_object = open('commodity_prices.txt', 'a')
file_object.write('\n'.join(commodity_list))
file_object.close()

# Uncomment below to fetch data from database table
# fetch_data()
```
